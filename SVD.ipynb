```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("bfi.csv")
df.info()
```

```python
df.drop(['rownames','gender','education','age'], axis=1, inplace=True)
```

```python
df.dropna(inplace=True)
```

```python
df
```

- $ U $ is eigen vectors of $ A  A^T $
- $ V $ is eigen vectors of $ A^T A $
- $ S $ is singular values $( \sqrt \lambda )$ 

```python
U, sigma, Vt = np.linalg.svd(df)
```

```python
sigma
```

#### Checking if we can remake the original matrix using U, Sigma, Vt

```python
df.shape, U.shape, sigma.shape, Vt.shape
```

```python
new_sigma = np.zeros((2436, 25))
np.fill_diagonal(new_sigma, sigma)
print(df)
df_remake = (U @ new_sigma @ Vt)
df_remake = pd.DataFrame(df_remake)
print(df_remake)
```
